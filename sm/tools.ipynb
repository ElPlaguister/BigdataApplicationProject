```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('서울시_cp949.csv', encoding = 'cp949')
df.head()
```

```python
df = df.drop(columns = ['가격', '매매호가', '전세호가', '월세호가', '실거래가'])
```

```python
df = df.drop(columns = ['용적률', '건폐율', '전용율', '임대세대수'])
```

```python
df.columns
```

```python
df['주차비율_정수'] = df['세대수'] // df['주차대수']
```

```python
df['주차비율_실수'] = df['세대수'] / df['주차대수']
```
