```python
API_key = "AIzaSyAemzn9ziZ9UQAqdrPJhQlCJx7dLzaWpiw"
gmaps = googlemaps.Client(API_key)

def outputer(loc_name):
    geocode_result = gmaps.geocode((loc_name), language='ko')
    try:
        latitude  = geocode_result[0]["geometry"]["location"]["lat"] # 리스트에서 위도 추출
        longitude = geocode_result[0]["geometry"]["location"]["lng"] # 리스트에서 경도 추출
    
    except Exception as e:
        print(e)
        return 0.0, 0.0
    return latitude, longitude
```
